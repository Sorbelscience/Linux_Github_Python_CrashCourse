- Short for list, this command "lists" all files that are not hidden and folders in your current working directory.

### A tack flag allows you some control as the user to see things at a more discrete or robust level

# The rm command

## rm aka the remove command
- Pretty straight foward like the previous commands. This command removes files or directories from your system. Again, this command has no "are you sure?" promot so be very sure you want to remove your target. There is not an easy way to restore files if they are deleted via the command line.

## The flags -rf
- The flags 'r' and 'f' are exclusively used for directory deletion. Due to hidden file scripts that are created when you make a directory. We need to recursively delete all the contents starting at the lowest directory and delete back to the pwd. 'f' means force which allows you to bypass typing yes to confirm certain directory or file contents being deleted for a directory. Directories are the only exception to the rule about being prompted for deletion.
# Example of rm for files
```
$joker422:ls
Documents Pictures Desktop newFile.txt
$joker422:rm newFile.txt
$joker422:ls
Documents Pictures Desktop 
```

# Example of rm for directories
```
$joker422:ls
Documents Pictures Desktop Project

# We use -rf for directories b/c directories have hidden files and folders #when they are created that manage commands to navigate in and out of them
$joker422:rm -rf Project/

$joker422:ls
Documents Pictures Desktop 
```

# Copying and Renaming Files and Directories
## 'cp' and 'mv'

# 'cp' Copy Files
- To copy a file, use cp followed by the target file name and the new file name. In addtion you can specify the file's target destination.

# Files
```
# pwd
$joker422: cp text.txt text1.txt
# inside a different Directory
$joker422: cp text.txt Documents/text1.txt

```

# The same goes for directories
```
$ cp -R <source_folder> <destination_folder>
```

```
# pwd
$joker422: cp  Documents Documents1
# inside a different Directory
$joker422: cp Documents Documents/Documents1
```
# 'mv' move(aka rename)
- mv allows you to do something similar to copy. What we see is the file or directory being "renamed" but there are some Linux commands happening under the hood so to speak with this operation.

## To do one 'mv' command Linus executes some system commands you just learned in this this order:

- The original file is copied to a temp file (cp)
- The original file is deleted (rm)
- A new file is created with the name you chose (touch)
- The temp file is read and written to the new file 
- The temp file is removed (rm)

# Rename a File
```
# pwd
$joker422: mv text.txt text1.txt
# inside a different Directory
$joker422: mv text.txt Documents/text1.txt
```

# Rename a Directory
```
$joker422: mv  Documents Documents1
# inside a different Directory
$joker422: mv Documents/Homework Documents/Busy_work
```
